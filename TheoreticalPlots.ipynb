```python
import numpy as np
import matplotlib.pyplot as plt
from math import log10
```

```python
vertices = np.arange(100, 3001, 1)

matrix = vertices * vertices

log_vertices = np.array([log10(item) for item in vertices])
list_half = (vertices + (vertices - 1) * vertices / 2) * log_vertices
list_max = (vertices + (vertices - 1) * vertices) * log_vertices
list_third = (vertices + (vertices - 1) * vertices * 0.3) * log_vertices
list_ten = (vertices + (vertices - 1) * vertices * 0.1) * log_vertices
```

```python
plt.figure(figsize=(7, 7))
plt.title("Comparison of Time Complexities")
plt.xlabel("Vertices")
plt.ylabel("Iterations")

plt.plot(vertices, matrix, label="Adjacency Matrix")
plt.plot(vertices, list_ten, label="Adjacency List (10% Edges)")
plt.plot(vertices, list_third, label="Adjacency List (30% Edges)")
plt.plot(vertices, list_half, label="Adjacency List (50% Edges)")
plt.plot(vertices, list_max, label="Adjacency List (Maximum Edges)")

plt.tight_layout()
plt.legend();
```
